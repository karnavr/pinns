## **KdV**

### **Single Soliton Solutions**

```python
model = KDV(num_solitons=1, seed=1001)
losses = model.train()
```

```python
model.plot_losses(model.losses)
```

```python
model.plot_losses(model.losses, component=['pde', 'initial', 'boundary', 'total'])
```

```python
model.plot_spacetime()
```

```python
model.plot_profiles(t_points=[0.0, 1.0, 5.0, 10.0])
```

```python
model.test(plot_heatmap=True)
```

```python
model.save_model_result('plotting/results/kdv_1_soliton.json')
```
